```python
import pandas as pd
pd.set_option('display.max_columns', 500)
from IPython.display import display
pd.set_option('display.max_rows', 100)
import plotly.graph_objects as go
```

```python
# Load the Excel file
xls = pd.ExcelFile('../datasources/RAM/Excel/RAMLDB v4.61 (assessment data only).xlsx')

# Get the names of all sheets in the Excel file
sheet_names = xls.sheet_names

# Print the names of all sheets
# for name in sheet_names:
#     print(name)

# Load each sheet into a separate dataframe
dfs = {sheet_name: xls.parse(sheet_name) for sheet_name in sheet_names}
# Now, dfs is a dictionary where the keys are sheet names and the values are dataframes
```

```python
#display head of each file
for name, df in dfs.items():
    print(f"Displaying head for sheet: {name}")
    display(df.head())
```

```python
SBFT_timeseries_values = dfs['timeseries_values_views'][dfs['timeseries_values_views']['stockid']=='SBT']
SBFT_timeseries_values.head()
```

```python
#TBbest is total biomass best?
#SSB is spawning stock biomass
#R or recruitment is the amount of fish that reach maturity
df = SBFT_timeseries_values[['stockid', 'year', 'TB', 'R']]
df.head()
```

```python
# Create a line plot
fig = go.Figure()

# Add a line for TB
fig.add_trace(go.Scatter(
    x=df['year'],
    y=df['TB'],
    mode='lines',
    name='TB'
))

# Add a line for R
fig.add_trace(go.Scatter(
    x=df['year'],
    y=df['R'],
    mode='lines',
    name='R'
))

# Add title and labels
fig.update_layout(
    title='Fish Stock Data Over Time',
    xaxis_title='Year',
    yaxis_title='Value',
)

# Show the plot
fig.show()
```
